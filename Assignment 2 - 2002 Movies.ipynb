```python
#Credit to the Professor, Ted Spence for the following code to access meta critic
# capturing data from meta critic
#now trying capturing data using regex, pyhton
import re #library for regex
import urllib3
import certifi

# Demonstration of how to fetch text from a website using certifi, urllib3, and UTF-8 encoding.
# This also simulates a browser agent.
# Adapted from https://tedspence.com/teaching-python-and-jupyter-cbffb68e2194


# This line of code generates a pool for HTTP requests.  Pooling means that a second request to the same
# website will be faster than the first one since it will retain an open connection for a period of time.
http = urllib3.PoolManager(ca_certs=certifi.where())


# These lines of code choose which website we will fetch and which browser we will simulate.
url = 'https://www.metacritic.com/browse/movies/score/metascore/year/filtered?year_selected=2002&sort=desc&view=detailed'
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0"

# This instruction requests the contents of the page and decodes it using UTF-8
response = http.request('GET', url, headers={"User-Agent": user_agent})
page_content = response.data.decode('utf-8')


#Getting movie titles. The findall function is used to retrieve data because we want to search only for specific items within the converted page (page_content)
movie_title= re.findall("<h3>(.*)</h3></a>",page_content)


# This is to get the movie posters
movie_poster= re.findall("a href=\"\/movie\/.*<img src=\"(.*)\" alt=\".*", page_content)


#To get the movie release date
release_date= re.findall("<span>(?!.*href)(.*?)<\/span>", page_content)


#to get the movie meta score
meta_score=re.findall("<div class=\"metascore_w large movie positive\">(.*)\s*<\/div>\s*<\/a>\s*<\/div>\s*<div", page_content)


#to get the movie summary
summary= re.findall("<div class=\"summary\">\s*(.*?)\s*<\/div>", page_content)


#--------------------


#Displaying data in panda data frame, thanks to Michael Etelamaki
import pandas as pd


# Define column names
column_names = ["Movie Title", "Movie Thumbnail", "Movie Meta Score", "Summary of Movie"]


# Create a dictionary to store the lists
data = {
  "Title of Movie": [movie_title],
  "Movie Thumbnail": [movie_poster],
  "Movie Meta Score": [meta_score],
  "Movie Release Date": [release_date],
  "Summary of Movie": [summary]
}

data = {column_name: lst for column_name, lst in zip(column_names, [movie_title, movie_poster, meta_score, summary])}


# Create DataFrame from dictionary
df = pd.DataFrame(data)


# for the index to start from 1 instead of 0, thanks to Jooyen Park
df.index += 1 


#to make the table aligned to the right or left, you can use the function .set_table_styles()
# thanks to Geeks for Geeks: https://www.geeksforgeeks.org/align-columns-to-left-in-pandas-python/
left_aligned_df = df.style.set_properties(**{'text-align': 'left'})
left_aligned_df = left_aligned_df.set_table_styles(
[dict(selector = 'th', props=[('text-align', 'left')])])


# Show all the rows and columns
display(left_aligned_df)


```
